### ※ **예제 문제**
* 설계공간: 0~1
* 목적함수: y = (6x - 2)<sup>2</sup> sin(12x - 4)의 최솟값 찾기
* 제한조건: 없음
* 다양한 초기값과 최적화 알고리즘을 적용하여 결과 비교해보기
* 정해: x<sub>opt</sub> = 0.75724875, f<sub>opt</sub> = -6.02074006

```python
from scipy.optimize import minimize, differential_evolution
import numpy as np
import matplotlib.pyplot as plt
```

```python
def obj(x): # 목적함수
    return ((6*x-2)**2)*np.sin((6*x-2)*2)
```

```python
def constr(x): # 제한조건함수
    return 0
```

```python
bnds = [(0,1)] # 경계값
```

```python
ineq_cons = {'type': 'ineq', 'fun': constr} # 제한조건 설정
```

```python
initial_points = np.array([0.0], dtype = 'float64') # 초기값
```

```python
print("Initial input",initial_points)
print("Initial obj",obj(initial_points))
```

```python
# 최적화 수행
result = minimize(obj,initial_points,method='SLSQP',bounds=bnds,constraints=ineq_cons) # Gradient-based optimization
# result = differential_evolution(obj,bnds) # Metaheuristic optimization
```

```python
result
```

```python
x_opt = result.x
f_opt = result.fun
# constr(result.x)
print("Optimal input",x_opt)
print("Minimum",f_opt)
```

```python
# 시각화 (Post-processing)
num = 100
x = np.linspace(0.0, 1.0, num) # 0에서 1까지 등간격으로 num개의 x값을 생성

y_true = obj(x)

# add a plot with variance
fig, axs = plt.subplots(1)
axs.plot(x, y_true) # 실제함수
axs.plot(x_opt, f_opt, "*") # 최적해

axs.set_xlabel("x")
axs.set_ylabel("y")
axs.legend(
    ["True function","Optimum"],
    loc="best",
)
plt.show()
```
