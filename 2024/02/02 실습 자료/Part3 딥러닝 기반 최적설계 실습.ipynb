```python
import numpy as np
from pyDOE import lhs
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import csv

# Number of samples
num_samples = 100

# Number of variables
num_variables = 11

# Min and max range of design variables
min_range = 0.001
max_range = 0.0145
```

```python
# pyDOE를 이용하여 Latin Hypercube Samples 데이터 조합을 만드는 코드!
lhs_samples = lhs(num_variables, samples=num_samples)
```

```python
# 통계적 이론 하에 원하는 구간 샘플링!
samples = min_range + lhs_samples * (max_range - min_range)

# 최소한의 데이터 조합으로 다양한 통계적 특성을 얻도록 보장하는
# Cumulative Distribution Function (CDF) 계산 코드!

cdf_samples = np.sort(samples, axis=0)
cdf_values = np.arange(1, num_samples + 1) / num_samples
```

# Visualization (시각화)

```python
# 시각화를 위한 화면 정의
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_title('Generated Design Spaces with CDF')

# 위에서 계산된 (샘플링된) 변수 조합 시각화
for i in range(num_variables):
    ax.scatter(samples[:, i], [i] * num_samples, cdf_values, label=f'Variable {i+1}', alpha=0.8)

ax.set_xlabel('Variable Values')
ax.set_ylabel('Variable Index')
ax.set_zlabel('CDF')
ax.legend(loc='upper right')

plt.show()
```

```python
# 생성된 디자인 스페이스와 CDF 값을 CSV 파일에 저장!
with open('design_spaces1.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Variable'] + [f'Var{i+1}' for i in range(num_variables)] + ['CDF'])
    for i in range(num_samples):
        csvwriter.writerow([i] + list(samples[i]) + [cdf_values[i]])
# 해당 값을 Ansys parameteric study 혹은 Abaqus와 연동하면 자동화 해석을 할 수 있음!
```

# 입력(환기 설비 변수) - 출력(오염물 농도) 사이 관계 예측 모델링 -->
# Deep neural network (DNN)!

![슬라이드2 - 복사본](https://github.com/DonggeunPark/DG/assets/131414228/4b0b8eb1-80dd-49d2-a763-e5cb989720f6)

```python
# 첫번째 block 정의
# Dense : 설계 변수 x1,x2,x3가 32개의 neuron 연산자로 전달
# 수학적 의미: Y_i = W * X, W는 최적화 시킬 파라미터, X는 입력 설계 변수를 의미,
# Y_i는 i번째 block에서 결과값이고 i+1 block에 새로운 변수로 입력됌

inputs = Input(shape=(3,)) # 3개의 입력 변수를 "inputs"이라는 변수로 지정함.
block1 = Dense(6, activation='relu')(inputs) # 3개의 입력 변수를 32개의 뉴런으로 연결하여 3*32개의 가중치를 갖고 relu 활성화 함수를 연결함.
block1 = BatchNormalization()(block1)  # 32개의 뉴런에 BatchNormalization 적용하여 오버피팅 문제 방지함.
block1 = Dropout(0.1)(block1)  # 32개 뉴런 중 10% dropout시켜 오버피팅 문제를 방지함
```

![슬라이드4](https://github.com/DonggeunPark/DG/assets/131414228/992cb25e-9ced-4fe1-8cfd-af4e11c67dff)

![슬라이드2 - 복사본](https://github.com/DonggeunPark/DG/assets/131414228/4b0b8eb1-80dd-49d2-a763-e5cb989720f6)


```python
block2 = Dense(12, activation='relu')(block1) # block 1 (Y_i)가 2번째 hidden layer 로 전달하여 64개 뉴런 생성 32*64개 학습 가중치 생성
block2 = BatchNormalization()(block2)
block2 = Dropout(0.1)(block2)

block3 = Dense(6, activation='relu')(block2) #block 2- 3 hiden layer 연결
block3 = BatchNormalization()(block3)
block3 = Dropout(0.1)(block3)

block4 = Dense(2, activation='relu')(block3)#block 3- 4 hiden layer 연결
#block4 = BatchNormalization()(block4)
block4 = Dropout(0.1)(block4)
```

```python
# 출력부 정의
outputs = Dense(2)(block4) #예측하고자 하는 objective function 개수 만큼 "output" 변수에 정의
```

# 딥러닝 모델화
#### * 앞서 구축한 deep learning architecture를 딥러닝 모델화를 통해 학습 파라미터 (W) 역전파 준비

```python
# 딥러닝 모델을 구축하기 위한 model 함수 호출
from tensorflow.keras.models import Model

model = Model(inputs=inputs, outputs=outputs)
#앞서 정의된 inputs과 outputs은 여러가지 layer를 동반한 block을 연속적으로 연결하고 있으며 Model 함수로 캡슐화함.
#즉, model은 현재 많은 가중치를 학습이 안된 상태로 랜덤하게 분포하고 이를 데이터에 맞게 최적화 시키는 과정을 "딥러닝 학습" 이라함.
```

# 딥러닝 학습 파라미터 (가중치) 역전파 학습을 위한 정의 부분

![슬라이드5](https://github.com/DonggeunPark/DG/assets/131414228/4ed8770a-1c44-44f9-bf91-26ffe6d182ca)

```python
from tensorflow.keras.optimizers import Adam
learning_rate = 0.0008
# 학습률에 따라 위의 정의된 딥러닝 모델을 적절히 loss가 감소되도록 만들며 이를 변경하는 것이 hyperparameter최적화라함.
opmization_variable = Adam(learning_rate=learning_rate)
# Compile the model
model.compile(optimizer=opmization_variable, loss='mean_squared_error')
#앞서 캡슐화된 model을 optimization algorithm과 연결하는 과정임.
```

```python
#학습된 모델 호출하여 최적화 알고리즘 연산에 사용함
import tensorflow as tf
json_file = open("dl.json", "r")
loaded_model_json = json_file.read()
json_file.close()
model = tf.keras.models.model_from_json(loaded_model_json)
model.load_weights("dl.h5")

# 목적함수 정의: 학습된 딥러닝 모델이 주어진 x로 부터 예측하는 오염원 농도
def f(x):
    pre = model.predict(x) #x 입력을 받아서, 모델이 pre 변수로 예측 값을 저장함!
    return pre[:,0]+pre[:,1]

def enforce_constraints(positions):
    # Apply constraints to keep positions within valid range
    positions[0] = np.clip(positions[0], 0.3, 2.5)
    positions[1] = np.clip(positions[1], 0.3, 2.5)
    return positions
```

```python
def update():
    "Function to do one iteration of particle swarm optimization"
    global V, input_position, pbest, pbest_obj, gbest, gbest_obj
    # Update params

    r1, r2 = np.random.rand(2) # random number
    V = w * V + c1*r1*(pbest - input_position) + c2*r2*(gbest.reshape(-1,1)-input_position) #brodcast
    # update V
    input_position = input_position + V
    # fitness calculation of updated swarm
    input_position = enforce_constraints(input_position)

    obj = f(np.vstack((input_position[0],input_position[1],np.array([[4]*30]))).T) # *5 is partivle number
    #pbest updata!
    pbest[:, (pbest_obj[:] >= obj[:])] = input_position[:, (pbest_obj[:] >= obj[:])]
    # if updated X is better that previous pbest personal best,
    # update the personel best to the updated X
    pbest_obj = np.array([pbest_obj[:], obj[:]]).min(axis=0)
    # pbest_obj와 obj 중 더 작은애들 업데이트

    gbest = pbest[:, pbest_obj[:].argmin()]
    #global update
    gbest_obj = pbest_obj.min()

def figure():
    update() #반복해서 시각화를 위함!
    fig, ax = plt.subplots(figsize=(8*0.5,6*0.5), dpi = 100)
    fig.set_tight_layout(True)
    x_pred = np.linspace(0.03, 2.5, 100)   # range of porosity values
    y_pred = np.linspace(0.03, 2.5, 100)
    xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)

    z_pred = np.ones(100*100)*4

    model_input = np.array([xx_pred.flatten(),yy_pred.flatten(),z_pred])

    z = f(model_input.T)
    img = ax.imshow(z[:].reshape(100,100), extent=[0.03, 2.5, 0.03,2.5], origin='lower', cmap='YlGnBu', alpha=0.5)
    # z is surface calculated by fitness function

    fig.colorbar(img, ax=ax)
    # ax.plot([x_min], [y_min], marker='x', markersize=5, color="white")
    # global optimum
    x, y = np.array(np.meshgrid(np.linspace(0.03, 2.5, 100), np.linspace(0.03, 2.5, 100)))

    contours = ax.contour(x, y, z[:].reshape(100,100), 10, colors='black', alpha=0.4)
    ax.clabel(contours, inline=True, fontsize=8, fmt="%.2f")
    pbest_plot = ax.scatter(pbest[0], pbest[1], marker='o', color='black', alpha=0.5)
    #각 변수들에 대한 pbest
    p_plot = ax.scatter(input_position[0], input_position[1], marker='o', color='blue', alpha=0.5)
    p_arrow = ax.quiver(input_position[0], input_position[1], V[0], V[1], color='blue', width=0.005, angles='xy', scale_units='xy', scale=1)
    gbest_plot = plt.scatter([gbest[0]], [gbest[1]], marker='*', s=100, color='black', alpha=0.4)
    ax.set_xlim([0.03,2.5])
    ax.set_ylim([0.03,2.5])
```
